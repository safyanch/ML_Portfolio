```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/titanic_clean.csv')
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Nulls:", df.isnull().sum().sum())
df.head()
```

```python
x=df.drop('Survived', axis=1)
y=df['Survived']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=.2, random_state=20)
```

```python
lr_model1 = LogisticRegression(max_iter=1000)
lr_model1.fit(x_train, y_train)
lr_pred1= lr_model1.predict(x_test)
```

```python
lr_model2 = RandomForestClassifier()
lr_model2.fit(x_train, y_train)
lr_pred2= lr_model2.predict(x_test)
```

```python
lr_model3 = SVC()
lr_model3.fit(x_train, y_train)
lr_pred3= lr_model3.predict(x_test)
```

```python
# Evaluation
```

```python
# Model 1 : logistic Regression
print("Logistic Regression")
print(accuracy_score(y_test, lr_pred1))
print(f1_score(y_test, lr_pred1))
print(confusion_matrix(y_test, lr_pred1))
```

```python
# Model 2 : Random Forest
print("Random Forest")
print(accuracy_score(y_test, lr_pred2))
print(f1_score(y_test, lr_pred2))
print(confusion_matrix(y_test, lr_pred2))
```

```python
# Model 3 : Support Vector Machine
print("Support Vector Machine")
print(accuracy_score(y_test, lr_pred3))
print(f1_score(y_test, lr_pred3))
print(confusion_matrix(y_test, lr_pred3))
```
